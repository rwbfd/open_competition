This is a demo for using CatBoost library. The reason why this is composed as a library instead of a script is because the interface is relatively simple, therefore it is easier to use as a code snippet instead of a script. 

Required the following libraries to run
1. lightgbm
2. numpy (or pandas)

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import catboost as cb
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
X, y = cancer.data,cancer.target
train_x, test_x, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=0)
 
```

```python
cat_features_index = [0,1,2,3,4,5,6]
 
def auc(m, train, test):
    return (metrics.roc_auc_score(y_train,m.predict_proba(train)[:,1]),
                            metrics.roc_auc_score(y_test,m.predict_proba(test)[:,1]))
 
```

```python
params = {'depth': [4, 7, 10],
          'learning_rate' : [0.03, 0.1, 0.15],
         'l2_leaf_reg': [1,4,9],
          'border_count':[20,25,30],
          'ctr_border_count':[40,50,60],
         'iterations': [300]}
model = cb.CatBoostClassifier()
cb_model = GridSearchCV(model, params, scoring="roc_auc", cv = 5)
cb_model.fit(train_x, y_train)

#cb_model.grid_scores_, gsearch1.best_params_, gsearch1.best_score_
```

```python
from sklearn.metrics import roc_auc_score

y_pred = model.predict(test_x)
print(roc_auc_score(y_test, y_pred))
```
